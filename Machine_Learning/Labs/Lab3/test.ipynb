```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 隨機生成一個類別內差異較小的數據集
np.random.seed(2)
class_0 = np.random.normal(loc=0, scale=0.2, size=(50, 2))
class_1 = np.random.normal(loc=1, scale=0.2, size=(50, 2))

# 計算均值
mean_0 = np.mean(class_0, axis=0)
mean_1 = np.mean(class_1, axis=0)

# 計算中點
middle_point = (mean_0 + mean_1) / 2

# 計算投影方向
direction = mean_1 - mean_0
direction /= np.linalg.norm(direction)  # 標準化

# 投影到一條線上
projected_0 = class_0 @ direction
projected_1 = class_1 @ direction

# 繪製原始數據集
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(class_0[:, 0], class_0[:, 1], color='blue', alpha=0.5, label='Class 0')
plt.scatter(class_1[:, 0], class_1[:, 1], color='red', alpha=0.5, label='Class 1')
plt.title('Same Performance')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.axvline(x=(mean_0 + mean_1) / 2 @ direction, color='green', linestyle='--', label='Decision Boundary')

# 標註均值和中點
plt.scatter(mean_0[0], mean_0[1], color='blue', marker='X', s=100, label='Mean Class 0 (m1)')
plt.scatter(mean_1[0], mean_1[1], color='red', marker='X', s=100, label='Mean Class 1 (m2)')
plt.scatter(middle_point[0], middle_point[1], color='purple', marker='o', s=100, label='Middle Point')
plt.legend()
plt.grid()

# 繪製各自的 Gaussian 分佈
plt.subplot(1, 2, 2)
x = np.linspace(-1, 3, 100)

# 計算 Class 0 的 PDF
mean_projected_0 = np.mean(projected_0)
std_projected_0 = np.std(projected_0)
pdf_0 = norm.pdf(x, mean_projected_0, std_projected_0)

# 計算 Class 1 的 PDF
mean_projected_1 = np.mean(projected_1)
std_projected_1 = np.std(projected_1)
pdf_1 = norm.pdf(x, mean_projected_1, std_projected_1)

# 繪製 PDF
plt.plot(x, pdf_0, color='blue', label='Gaussian Class 0', lw=2)
plt.plot(x, pdf_1, color='red', label='Gaussian Class 1', lw=2)
plt.title('Gaussian Distributions of Projected Data')
plt.xlabel('Projected Feature')
plt.ylabel('Probability Density')
plt.axvline(x=mean_projected_0, color='blue', linestyle=':', label='Mean Class 0 (m1)')
plt.axvline(x=mean_projected_1, color='red', linestyle=':', label='Mean Class 1 (m2)')
plt.axvline(x=middle_point @ direction, color='purple', linestyle=':', label='Middle Point')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 設定隨機種子以便重現
np.random.seed(42)

# 生成均勻分布的數據集
class_0 = np.random.uniform(low=-1, high=0, size=(100, 2))  # 第一類數據
class_1 = np.random.uniform(low=0, high=1, size=(100, 2))   # 第二類數據

# 繪製數據集
plt.figure(figsize=(8, 6))
plt.scatter(class_0[:, 0], class_0[:, 1], color='blue', alpha=0.5, label='Class 0')
plt.scatter(class_1[:, 0], class_1[:, 1], color='red', alpha=0.5, label='Class 1')

# 計算均值
mean_0 = np.mean(class_0, axis=0)
mean_1 = np.mean(class_1, axis=0)

# 計算中點
middle_point = (mean_0 + mean_1) / 2

# 繪製均值和中點
plt.scatter(mean_0[0], mean_0[1], color='blue', marker='X', s=100, label='Mean Class 0 (m1)')
plt.scatter(mean_1[0], mean_1[1], color='red', marker='X', s=100, label='Mean Class 1 (m2)')
plt.scatter(middle_point[0], middle_point[1], color='purple', marker='o', s=100, label='Middle Point')

# 繪製決策邊界
plt.axvline(x=middle_point[0], color='cyan', linestyle='--', label='Decision Boundary')

# 設置標題和標籤
plt.title('Uniform Distribution Dataset Example')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid()
plt.xlim(-1.5, 1.5)
plt.ylim(-0.5, 1.5)

# 顯示圖形
plt.tight_layout()
plt.show()
```
