```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PIR_01 = []
with open('Dyna_0.01','r') as file:
    for line in file:
        PIR_01.append(line.strip())
PIR_015 = []
with open('Dyna_0.015','r') as file:
    for line in file:
        PIR_015.append(line.strip())
PIR_02 = []
with open('Dyna_0.02','r') as file:
    for line in file:
        PIR_02.append(line.strip())
PIR_1 = []
with open('Dyna_0.1','r') as file:
    for line in file:
        PIR_1.append(line.strip())
```

```python
PIR_01 = np.float32(np.array(PIR_01))
PIR_015 = np.float32(np.array(PIR_015))
PIR_02 = np.float32(np.array(PIR_02))
PIR_1 = np.float32(np.array(PIR_1))
```

```python
Packet_size = [2,4,8,16,32,48,64]
```

```python
colors = ['#1f77b4','#ff7f0e','#2ca02c','#be5b80']

plt.figure(figsize=(10,6))

plt.plot(Packet_size,PIR_01,label='0.01 PIR',color=colors[0],marker='o')
plt.plot(Packet_size,PIR_015,label='0.015 PIR',color=colors[1],marker='o')
plt.plot(Packet_size,PIR_02,label='0.02 PIR',color=colors[2],marker='o')
plt.plot(Packet_size,PIR_1,label='0.1 PIR',color=colors[3],marker='o')

plt.title('Energy used')
plt.xlabel('Packet size')
plt.ylabel('Dynamic energy')
plt.legend()

plt.show()
```
